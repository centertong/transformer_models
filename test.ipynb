```python
import torch
import einops
```

```python
q = torch.tensor([1,2,2])
k = torch.tensor([2,1])
```

```python
qr = einops.repeat(q, 'd -> d b', b = 2)
kr = einops.repeat(k, 'd -> b d', b = 3)
```

```python
sm = qr == kr
```

```python
qv = torch.rand((3, 4))
kv = torch.rand((2,4))
```

```python
qv
```

```python
sq = torch.einsum('qd, qk -> qkd', qv, sm)
```

```python
sk = torch.einsum('kd, qk -> kqd', kv, sm)
```

```python
torch.einsum('qkd, kqd-> qk', sq, sk)
```
